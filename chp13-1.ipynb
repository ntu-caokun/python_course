## 13.2.2 scipy简单应用
二维图像卷积运算。

```python
import numpy as np
from scipy import signal
from scipy import misc
import matplotlib.pyplot as plt

# face = misc.face(gray=True) # misc.face 现已被移至 scipy.datasets
import scipy.datasets
face = scipy.datasets.face(gray=True)
scharr = np.array([[ -3-3j, 0-10j, +3-3j],
                   [-10+0j, 0+0j, +10+0j],
                   [ -3+3j, 0+10j, +3+3j]]) # Gx + j*Gy
grad = signal.convolve2d(face, scharr, boundary='symm', mode='same')
```

## 13.2.2 scipy简单应用
二维图像卷积运算绘图

```python
fig, (ax_orig, ax_mag) = plt.subplots(1, 2, figsize=(10,6))
ax_orig.imshow(face, cmap='gray')
ax_orig.set_title('Original')
ax_orig.set_axis_off()
ax_mag.imshow(np.absolute(grad), cmap='gray')
ax_mag.set_title('Gradient magnitude')
ax_mag.set_axis_off()
fig.show()
```

## 13.2.2 scipy多重积分
$I=\int_{y=0}^{1/2}\int_{x=0}^{1-2y}xy~dx~dy=\frac{1}{96}.$

```python
from scipy import integrate
def f(x, y):
    return x*y
def bounds_y():
    return [0, 0.5]
def bounds_x(y):
    return [0, 1-2*y]
print(integrate.nquad(f, [bounds_x, bounds_y]))
```

## 13.2.2 scipy优化
最优化函数 (Rosenbrock)

```python
import numpy as np
from scipy.optimize import minimize
def rosen(x):
    """The Rosenbrock function"""
    return sum(100.0*(x[1:]-x[:-1]**2.0)**2.0 + (1-x[:-1])**2.0)
x0 = np.array([1.3, 0.7, 0.8, 1.9, 1.2])
res = minimize(rosen, x0, method='nelder-mead', options={'xatol': 1e-8, 'disp': True})
print(res.x)
```

## 13.2.2 scipy全局优化
Eggholder 函数 3D 可视化

```python
def eggholder(x):
    return (-(x[1]+47)*np.sin(np.sqrt(abs(x[0]/2+(x[1]+47)))) - x[0]*np.sin(np.sqrt(abs(x[0]-(x[1]+47)))))
bounds = [(-512, 512), (-512, 512)]
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
x = np.arange(-512, 513)
y = np.arange(-512, 513)
xgrid, ygrid = np.meshgrid(x, y)
xy = np.stack([xgrid, ygrid])
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.view_init(45, -45)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('eggholder(x, y)')
ax.plot_surface(xgrid, ygrid, eggholder(xy), cmap='terrain')
plt.show()
```

## 13.2.2 scipy全局优化
不同优化算法测试

```python
from scipy import optimize
results = dict()
results['DA'] = optimize.dual_annealing(eggholder, bounds)
print(results['DA'])
results['DE'] = optimize.differential_evolution(eggholder, bounds)
results['BH'] = optimize.basinhopping(eggholder, bounds)
```

## 13.2.2 scipy全局优化
优化结果 2D 可视化标记

```python
fig = plt.figure()
ax = fig.add_subplot(111)
im = ax.imshow(eggholder(xy), interpolation='bilinear', origin='lower', cmap='gray')
ax.set_xlabel('x')
ax.set_ylabel('y')
def plot_point(res, marker='o', color=None):
    ax.plot(512+res.x[0], 512+res.x[1], marker=marker, color=color, ms=10)
plot_point(results['BH'], color='y')
plot_point(results['DE'], color='c')
plot_point(results['DA'], color='w')
ax.set_xlim([-4, 514*2])
ax.set_ylim([-4, 514*2])
plt.show()
```
