# 第11章 异常处理结构与单元测试

异常是指程序运行时引发的错误，如除零、下标越界、文件不存在等。合理使用异常处理可以使程序更健壮。

## 11.1 异常处理结构

### 常见的异常表现形式：
* `ZeroDivisionError`: 除零错误
* `TypeError`: 类型不支持或参数类型不匹配
* `NameError`: 变量名不存在
* `FileNotFoundError`: 文件或目录不存在

### 语法结构
1. **try...except...**
2. **try...except...else...** (没有引发异常时执行else)
3. **try...except...finally...** (无论是否发生异常，都会执行finally)

```python
def div(x, y):
    try:
        print(x / y)
    except ZeroDivisionError:
        print('ZeroDivisionError: 不能除以0')
    except TypeError:
        print('TypeError: 类型错误')
    else:
        print('No Error')
    finally:
        print("executing finally clause\n")

div(3, 5)
div('3', 5)
div(3, 0)
```

### 11.1.4 断言与上下文管理语句
* **断言 `assert expression[, reason]`**: 当表达式为假时，抛出 `AssertionError`。
* **with 语句**: 无论发生什么异常，自动释放资源。

```python
a = 3
b = 5
try:
    assert a == b, 'a must be equal to b'
except AssertionError as reason:
    print('%s: %s' % (reason.__class__.__name__, reason))
```

## 11.2 单元测试

单元测试是保证模块质量的重要手段，Python标准库 `unittest` 提供了大量用于单元测试的类和方法。

**常用断言方法：**
| 方法名称 | 功能说明 |
|---|---|
| `assertEqual(a, b)` | `a == b` |
| `assertTrue(x)` | `bool(x) is True` |
| `assertIs(a, b)` | `a is b` |
| `assertIsNone(x)` | `x is None` |
| `assertIn(a, b)` | `a in b` |
| `assertIsInstance(a, b)` | `isinstance(a, b)` |
| `setUp()` | 每项测试开始前自动调用 |
| `tearDown()` | 每项测试完成后自动调用 |

```python
import unittest

class TestStringMethods(unittest.TestCase):

    def test_upper(self):
        self.assertEqual('foo'.upper(), 'FOO')

    def test_isupper(self):
        self.assertTrue('FOO'.isupper())
        self.assertFalse('Foo'.isupper())

    def test_split(self):
        s = 'hello world'
        self.assertEqual(s.split(), ['hello', 'world'])
        # 检查是否抛出 TypeError
        with self.assertRaises(TypeError):
            s.split(2)

# 在 Jupyter Notebook 中运行 unittest 需要一点特殊参数
# if __name__ == '__main__':
#     unittest.main(argv=['first-arg-is-ignored'], exit=False)
```
